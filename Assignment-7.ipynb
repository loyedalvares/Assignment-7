```python
# 1. Create an sqlalchemy engine using a sample from the data set
```

```python
import sqlalchemy as db
import pandas as pd

engine = db.create_engine('sqlite:///adultData.db')
connection = engine.connect()
alc_adult_data.to_sql('adultAllData',connection, if_exists='replace')
metadata = db.MetaData()
adults_data = db.Table('adultAllData', metadata, autoload=True, autoload_with=engine)

# Fetching the records from the table
query = db.select([adults_data])
result = connection.execute(query)
adlt_data = result.fetchall()
adlt_data
```

```python
# 4. Write two filter queries
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
Session = sessionmaker()
Session.configure(bind=engine) 
session = Session()
```

```python
# Filter Query 1
from sqlalchemy import and_

result = session.query(adults_data).filter(and_(adults_data.columns.education=='Doctorate', adults_data.columns.native_country.like('%States%'))).all()

for row in result:
    print()
```

```python
# Filter Query 2

from sqlalchemy import or_

result = session.query(adults_data).filter(or_(adults_data.columns.education=='Doctorate', adults_data.columns.education=='Masters')).all()

for row in result:
    print(row)
```

```python
# 5. Write two function queries
```

```python
# Function Query 1

from sqlalchemy import func
from sqlalchemy.sql import label

result = session.query(adults_data.columns.occupation,
                    label('Average_Age', func.avg(adults_data.columns.age)),
                    label('Minimum_Age', func.min(adults_data.columns.age)),
                    label('Maximum_Age', func.max(adults_data.columns.age))
                  ).group_by(adults_data.columns.occupation).all()

for row in result:
    print(row)
```

```python
# Function Query 2

result = session.query(adults_data.columns.occupation,
                    label('Count', func.count(adults_data.columns.education)),
                    label('Hours_Per_Week', func.sum(adults_data.columns.hours_per_week)),
                    label('Capital_Gain', func.sum(adults_data.columns.capital_gain)),
                    label('Capital_Loss', func.sum(adults_data.columns.capital_loss))
                  ).group_by(adults_data.columns.occupation).all()

for row in result:
    print(row)
```
